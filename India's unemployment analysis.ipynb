```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='latitude',data=df,palette='bright')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha="right")
plt.title('Distribution of number of unemployed people as per the latitude',size=20)
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='longitude',data=df,palette='bright')
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha="right")
plt.title('Distribution of number of unemployed people as per the longitude',size=20)
plt.ylabel('Frequency')
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['est_unemp_rate', 'est_emp', 'est_lab_part_rate']])
plt.title("Box Plot of Employment Data")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```

```python
sns.pairplot(df[['est_unemp_rate', 'est_emp', 'est_lab_part_rate']])
plt.title("Pairwise Scatter Plot")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='frequency', y='est_unemp_rate', data=df)
plt.title("Violin Plot of Employment Rate by Frequency")
plt.xlabel("Frequency")
plt.ylabel("Estimated Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='region', y='est_unemp_rate', data=df)
plt.title("Average Estimated Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Average Estimated Unemployment Rate")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='region', y='est_unemp_rate', data=df, inner=None)
sns.swarmplot(x='region', y='est_unemp_rate', data=df, color='k', alpha=0.7)
plt.title("Distribution of Estimated Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Estimated Unemployment Rate")
plt.xticks(rotation=60)
plt.show()
```

```python
avg_emp_by_state = df.groupby('region')['est_emp'].mean()

plt.figure(figsize=(8, 8))
plt.pie(avg_emp_by_state, labels=avg_emp_by_state.index, autopct='%1.1f%%', startangle=140)
plt.title("Average Employment Rate by State")
plt.show()
```

```python
fig = px.sunburst(df, path=['region1', 'region'], values='est_emp')
fig.update_layout(title="Sunburst Plot of Employment by Region and Subregion", width=1200, height=600)
fig.show()
```

```python
fig = px.scatter_geo(df, 'latitude', 'longitude', scope='asia', hover_name='region', color = 'region', size = 'est_unemp_rate', animation_frame='Month')
fig.update_layout(title="Geographic Scatter Plot of Impact of Pandemic in India")
fig.update_geos(lataxis_range=[5,35], lonaxis_range=[65, 100], showocean=True)
fig.show()
```

```python
df['date'] = pd.to_datetime(df['date'])

# Extract the month from the 'date' column and create a new column 'month'
df['Month'] = df['date'].dt.month

# Calculate the average unemployment rates before and after pandemic
avg_unemp_before_pandemic = df[(df['Month'] >= 1) & (df['Month'] <= 4)]['est_unemp_rate'].mean()
avg_unemp_after_pandemic = df[(df['Month'] >4) & (df['Month'] <=12)]['est_unemp_rate'].mean()

print("Average Unemployment Rate Before Pandemic:", avg_unemp_before_pandemic)
print("Average Unemployment Rate After Pandemic:", avg_unemp_after_pandemic)
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(data=df, x='region1', y='est_unemp_rate', hue='year', ci=None, palette = ["#e60049"])
plt.title("Effect of Pandemic on Unemployment Rates by Region")
plt.xlabel("Region1")
plt.ylabel("Unemployment Rate")
plt.legend(title="Year", loc='upper left', labels=['After 2020'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='region', y='est_unemp_rate', hue='year', ci=None, palette=["#00ffff"])
plt.title("Effect of Pandemic on Unemployment Rates by States")
plt.xlabel("Region1")
plt.ylabel("Unemployment Rate")
plt.legend(title="Year", loc='upper left', labels=['After 2020'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=['Before Pandemic', 'After Pandemic'], y=[avg_unemp_before_pandemic, avg_unemp_after_pandemic])
plt.title("Average Unemployment Rate Comparison (Before and After Pandemic)")
plt.ylabel("Average Unemployment Rate")
plt.show()
```
