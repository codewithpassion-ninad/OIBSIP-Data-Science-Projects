5. Modeling

```python
le = LabelEncoder()
le.fit(df['target'])
df['encode_target'] = le.transform(df['target'])
```

```python
X = df['message']
y = df['encode_target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
print(len(X_train), len(y_train))
print(len(X_test), len(y_test))
```

```python
# instantiate the vectorizer
vect = CountVectorizer()
vect.fit(X_train)
CountVectorizer()
# Use the trained to create a document-term matrix from train and test sets
X_train_dtm = vect.transform(X_train)
X_test_dtm = vect.transform(X_test)
```

```python
nb = MultinomialNB()

# Train the model
nb.fit(X_train_dtm, y_train)
```

```python
y_pred_class = nb.predict(X_test_dtm)
y_pred_prob = nb.predict_proba(X_test_dtm)[:, 1]
print(accuracy_score(y_test, y_pred_class))
confusion_matrix(y_test, y_pred_class)
print(classification_report(y_test, y_pred_class))
```

```python
max_words = 1000  # Maximum number of words to keep based on word frequency
max_sequence_length = 100  # Maximum sequence length for padding

tokenizer = Tokenizer(num_words=max_words, oov_token='<OOV>')
tokenizer.fit_on_texts(df['message'])

X = tokenizer.texts_to_sequences(df['message'])
X = pad_sequences(X, maxlen=max_sequence_length, padding='post', truncating='post')

# Encode the target variable
label_encoder = LabelEncoder()
df['target'] = label_encoder.fit_transform(df['target'])

# Split the df into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df['target'], test_size=0.2, random_state=42)
```

<a id = "lstm"></a>
6. LSTM

```python
# Build the LSTM model
model = Sequential()
model.add(Embedding(input_dim=max_words, output_dim=128, input_length=max_sequence_length))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
# Train the model
batch_size = 64
epochs = 10
history = model.fit(X_train, y_train, batch_size=batch_size, epochs=epochs, validation_split=0.2)

# Evaluate the model
accuracy = model.evaluate(X_test, y_test)[1]
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
